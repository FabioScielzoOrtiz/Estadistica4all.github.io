# Join data_frames using `merge`

```python
Ventas
```

```python
Clientes
```

```python
pd.merge(Ventas, Clientes, on='ClienteID', how='inner')  # equivalent to inner join
```

```python
pd.merge(Ventas, Clientes, on='ClienteID', how='outer')  # equivalent to outer join
```

```python
pd.merge(Ventas, Clientes, on='ClienteID', how='left')  # equivalent to left join
```

```python
pd.merge(Ventas, Clientes, on='ClienteID', how='right')  # equivalent to right join
```
