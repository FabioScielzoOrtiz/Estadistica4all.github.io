### Recoding the categorical variables to the standard encoding (0,1,2,3,...)

To recode the categorical variables to the standard encoding we will use sklearn.preprocessing

This tool recode recode the categorical variables to the standard encoding using an alphabetical order. So the category recode as 0 will be the category such that it's first letter is greater (or equal) than the rest following the alphabetical order.

So if we have a categorical variable $X_j$ such that $Range(X_j) = \lbrace low , medium , high \rbrace $ , then sklearn.preprocessing encode $high = 0$ , $low = 1$ , $medium = 2$ , because following the alphabet order we have $h > l > m$

```python
from sklearn.preprocessing import OrdinalEncoder

ord_enc = OrdinalEncoder()
```

**neighborhood**

```python
House_Price_Regression['neighborhood_recode'] = ord_enc.fit_transform(House_Price_Regression[['neighborhood']])
```

**quality**

```python
House_Price_Regression['quality_recode'] = ord_enc.fit_transform(House_Price_Regression[['quality']])
```
